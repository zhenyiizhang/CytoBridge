
# Installation

Currently, CytoBridge is in the development stage. You can download it directly from this repository for testing and contribution.

---

## Installation Methods

### First Method: Using GitHub

Clone the repository using the following command:
```
git clone https://github.com/zhenyiizhang/CytoBridge.git
```

#### Basic Usage

1. **Create a new Conda environment**:
    ```
    conda create -n CytoBridge python=3.10 ipykernel -y
    ```
2. **Activate the environment**:
    ```
    conda activate CytoBridge
    ```
3. **Navigate to the CytoBridge directory**:
    ```
    cd path_to_CytoBridge
    ```
4. **Install required packages**:
    ```
    pip install -r requirements.txt
    ```

---

### Second Method: Using PyPI

Install CytoBridge directly using pip:
```
pip install CytoBridge
```

**Note**: When using PyPI, all dependencies will be automatically installed, so there is no need to manually install them.
